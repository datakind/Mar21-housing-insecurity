```python
## Importing libraries and data

import plotly.express as px
import pandas as pd

evictions = pd.read_csv("nyc_evictions_geocoded.csv")
```

```python
## Cleaning Data

# create a clean datetime column
evictions['Date'] = pd.to_datetime(evictions['EXECUTED_DATE'])
# create new columns with Year, Month, Day data
evictions['Year'], evictions['Month'], evictions['Day'] = evictions['Date'].dt.year, evictions['Date'].dt.month, evictions['Date'].dt.day 

# removing one row from 1970
evictions.drop(evictions[evictions.Year > 2021].index, inplace=True)
```

```python
fig = px.histogram(evictions, x="Year", template="ggplot2")

fig.update_layout(
    title_text='NYC Evictions by Year', # title of plot
    xaxis_title_text='Year', # xaxis label
    yaxis_title_text='Total Evictions', # yaxis label
    bargap=0.2, # gap between bars of adjacent location coordinates
)

fig.show()
```

```python
# to visualize by year, use pandas to create a subset df 

by_month = (evictions.groupby(['Month', 'Year']).
    size()
   .reset_index(name='count'))

fig = px.line(by_month, x="Month", y="count", color='Year', template="ggplot2")


# graphed with plotly

fig.update_layout(
    title_text='NYC Evictions by Month', # title of plot
    xaxis_title_text='Month', # xaxis label
    yaxis_title_text='Total Evictions', # yaxis label
)

fig.show()
```
